#### Visualization:

```python
df = pd.DataFrame(np.random.rand(10,4), index = pd.date_range('1/1/2020', periods = 10), columns = list('ABCD'))
df
```

```python
df.plot()
```

#### BarPlot

```python
df1 = pd.DataFrame(np.random.rand(10,4), columns=['a','b','c','d'])
df1.plot(kind = 'bar')       
# (or)  df1.plot.bar()
```

```python
# to get stacked barplot
df1.plot(kind = 'bar', stacked = True) 
```

```python
# to get horizontal barplot, use barh

df1.plot(kind = 'barh', stacked = True)
```

#### Histograms:

```python
df2 = pd.DataFrame({'a':np.random.randn(1000)+1, 'b':np.random.randn(1000), 'c': np.random.randn(1000) - 1}, 
                   columns=['a', 'b', 'c'])
df2
```

```python
df2.plot(kind = 'hist', bins=20)
```

#### BoxPlot
- Boxplot can be drawn calling 
    - Series.plot.box() 
    - DataFrame.plot.box(), 
    - or DataFrame.boxplot() to visualize the distribution of values within each column.

```python
df3 = pd.DataFrame(np.random.rand(10, 5), columns=['A', 'B', 'C', 'D', 'E'])
df3
```

```python
df3.plot.box()
```

```python
df3.boxplot()
```

```python
df3.plot.area()
```

#### Scatterplot

```python
df4 = pd.DataFrame(np.random.rand(50, 4), columns=['a', 'b', 'c', 'd'])
df4.plot.scatter(x='a', y='b')
```

```python
# or follow this syntax
df4.plot(kind = 'scatter', x='a', y='b')
```

#### Pie chart

```python
df5 = pd.DataFrame(3 * np.random.rand(4), index=['a', 'b', 'c', 'd'], columns=['x'])
df5
```

```python
df5.plot.pie(subplots=True)
```
